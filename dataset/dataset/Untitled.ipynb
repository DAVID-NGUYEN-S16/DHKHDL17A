```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as pltc
all_colors = [k for k,v in pltc.cnames.items()]
```

```python
data = pd.read_csv('orders.csv')
data
```

## Clearning data 

```python
list_data = list(data.columns)
y = []
percent = []
for x in list_data:
    colors = all_colors
    y.append(np.sum([data[x].isnull()]))
    percent.append((np.sum(([data[x].isnull()]))/len(data[x]))*100.0)
    
plt.subplot(1,1,1)
plt.barh(list_data, y,height=0.8, color = colors)
stactic_cells_null = pd.DataFrame({'Name Item':list_data,"Quantily of cell null": y, 'percent':percent})
stactic_cells_null
```
