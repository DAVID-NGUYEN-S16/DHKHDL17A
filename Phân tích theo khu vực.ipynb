```python
import pandas as pd
import numpy as np
df = pd.read_csv("orders.csv")
df.dropna(inplace = True)
df.info()
```

# Các tiểu bang

```python
# Ép kiểu datetime cho cột "Order date" và thêm cột "Year", "Month", gán cột Profit=Profit*Quantity
pd.options.display.float_format = '{:,.1f}'.format
df=df.loc[1:9195,['Customer Name', 'Order Date', 'City', 'State', 'Category', 'Quantity', 'Profit']]
df['Order Date']=pd.to_datetime(df['Order Date'], infer_datetime_format=True)
df["Month"]=df["Order Date"].dt.month
df["Year"]=df["Order Date"].dt.year
df['Profit'] = (df["Profit"]*df["Quantity"]).astype(float)
df
```

```python
# Gom các tiểu bang theo từng tháng
pd.options.display.float_format = '{:,.1f}'.format
df_region_month=pd.DataFrame(df, columns=["City", "State", "Profit", "Month"])
df_region_month=df_region_month.groupby(["State", "Month"], as_index=False)["Profit"].sum()
df_region_month
```

```python
# Gom các tiểu bang theo từng năm
pd.options.display.float_format = '{:,.1f}'.format
df_region_year=pd.DataFrame(df, columns=["City", "State", "Profit", "Year"])
df_region_year=df_region_year.groupby(["State", "Year"], as_index=False)["Profit"].sum()
df_region_year
```

```python
#Gom các tiểu bang tổng 4 năm
df_region_total=pd.DataFrame(df, columns=["State", "Profit"])
df_region_total=df_region_total.groupby(["State"], as_index=False)["Profit"].sum()
df_region_total.sort_values(["Profit"], ascending=[False], inplace=True)
df_region_total
```

# Thành phố

```python
# Gom các thành phố theo từng tháng
# Có thành phố mua hàng ít quá không đủ 12 tháng
pd.options.display.float_format = '{:,.1f}'.format
df_city_year=pd.DataFrame(df, columns=["City", "State", "Profit", "Month"])
df_city_year=df_city_year.groupby(["City", "Month", "State"], as_index=False)["Profit"].sum()
df_city_year.head(60)
```

```python
# Gom các thành phố theo từng năm
pd.options.display.float_format = '{:,.1f}'.format
df_city_year=pd.DataFrame(df, columns=["City", "State", "Profit", "Year"])
df_city_year=df_city_year.groupby(["City", "State", "Year"], as_index=False)["Profit"].sum()
df_city_year.head(60)
```

```python
# Gom các thành phố theo tổng 4 năm
pd.options.display.float_format = '{:,.1f}'.format
df_city_year=pd.DataFrame(df, columns=["City", "State", "Profit"])
df_city_year.sort_values(["Profit"], ascending=[False], inplace=True)
df_city_year.head(60)
```
